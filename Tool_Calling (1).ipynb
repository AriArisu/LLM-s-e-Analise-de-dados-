##Exercício: Criando um Agente com Tool Calling para Responder Perguntas com Cálculos

Neste exercício, você irá usar a funcionalidade de tool calling (chamada de ferramentas) da OpenAI para construir um agente que responde perguntas que requerem cálculo matemático, chamando uma função Python definida por você.

Modelos de linguagem são muito bons em linguagem natural, mas nem sempre são precisos com cálculos ou lógica matemática. Ao utilizar o recurso de tool calling, podemos conectá-los a funções externas confiáveis — como uma calculadora Python, sistema de busca ou banco de dados — para obter respostas precisas.

Para te auxiliar, use a função eval(expressao) do python para resolver cálculos matemáticos.


```python
def calcular(expressao):
    try:
        # Avalia a expressão matemática de forma segura
        resultado = eval(expressao)
        return resultado
    except Exception as e:
        return f"Erro ao calcular a expressão: {str(e)}"
```

```python
tools = [{
    "type": "function",
    "function": {
        "name": "calcular",
        "description": "Calcula a expressão matemática fornecida.",
        "parameters": {
            "type": "object",
            "properties": {
                "expressao": {"type": "string"}
            },
            "required": ["expressao"],
            "additionalProperties": False
        },
        "strict": True
    }
}]
```

```python
eval("1+1")
```

```python
from openai import OpenAI
from openai import OpenAI
import json

client = OpenAI(api_key="")

messages = [{"role": "user", "content": "Qual é o resultado de 25 + 36?"}]

completion = client.chat.completions.create(
    model="gpt-4o-mini",  # Pode ser outro modelo
    messages=messages,
    tools=tools,
)

tool_call = completion.choices[0].message.tool_calls[0]
args = json.loads(tool_call.function.arguments)

result = calcular(args["expressao"])

messages.append(completion.choices[0].message)  # Chamada da função
messages.append({  # Resultado da função
    "role": "tool",
    "tool_call_id": tool_call.id,
    "content": str(result)
})

completion_2 = client.chat.completions.create(
    model="gpt-4o-mini",  # Pode ser outro modelo
    messages=messages,
    tools=tools,
)

print(completion_2.choices[0].message.content)
```
