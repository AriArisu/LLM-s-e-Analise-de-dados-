```python
msgs = [{'role': 'system', 'content': 'Responda com confiança como se fosse um bombeiro passando instruções.Só responda a perguntas relacionadas a como lidar com enchentes e assuntos relacionadas, como animais peçonhentos'}]

while True:
    pergunta = input("Como posso ajudar ?")
    prompt = f"Com base no contexto fornecido, responda a pergunta do usuário.\nContexto: {context}\n\nPergunta: {pergunta}\nResposta:"
    msgs.append({'role': 'user', 'content': pergunta})
    completion = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=msgs
    )
    resposta = completion.choices[0].message.content
    print(resposta)
    msgs.append({'role': 'assistant', 'content': resposta})
```
